10. A partir del dataset y las columnas, utilizar un modelo de Regresión Linear de ScikitLearn para encontrar el modelo que pueda explicar la correlación entre `precio` y `area`. (4 pts)

Considerando la siguiente ecuación
$Y = X*w + b$

$Y$: `valor del precio`
$X$: `tamaño del departamento en m2` 

$w$ y $b$: coeficientes 


```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df_depas = pd.DataFrame(data)

#Tratado de datos

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)

df_depas['precio'] = df_depas['precio'].apply(convertirSoles).astype(int)

df_depas['area'] = df_depas['area'].str.replace(' m2','',case=False).astype(float)

simpledf = df_depas[['precio', 'area']]

#modelo de regresion
X = simpledf['area'].values.reshape(-1, 1)
Y = simpledf['precio']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=27)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

plt.scatter(X_train, Y_train, color='green', label='Entrenamiento')
plt.scatter(X_test, Y_test, color='red', label='Prueba')
plt.plot(X_train, model.predict(X_train), color= 'yellow', label='Línea de regresión')
plt.xlabel('Área por m2')
plt.ylabel('Precio en S/.')
plt.legend()
plt.show()
```

```python
#Imprime tus parámetros w y b aquí

# Obtener el coeficiente (w)
w = model.coef_[0]

# Obtener la intersección (b)
b = model.intercept_

print(f"Coeficiente (w): {w}")
print(f"Intercepto (b): {b}")
```

11.  ¿Cómo varía los coeficientes `w` y `b` dependiendo del distrito? Seleccione 5 casos y analice `w` y `b`. (2 pts)

```python
# Crear un diccionario para almacenar los coeficientes de cada distrito
coeficientes_distritos = {}

# Seleccionar 5 distritos
distritos_seleccionados = ['La Victoria', 'Santiago De Surco', 'Miraflores', 'Lince', 'San Miguel']

# Colores para los puntos de regresión de cada distrito
colores_punto = ['orange', 'blue', 'red', 'purple', 'green']

# Colores para las líneas de regresión de cada distrito
colores_linea = ['darkorange', 'navy', 'darkred', 'indigo', 'darkgreen']

# Crear el gráfico
plt.figure(figsize=(12, 8))

# Recorrer cada distrito y crear un modelo de regresión para cada uno.
for i, distrito in enumerate(distritos_seleccionados):
    # Filtrar los datos
    df_distrito = df_depas[df_depas['distrito'] == distrito]
    
    # Definimos variables
    X = df_distrito['area'].values.reshape(-1, 1)
    Y = df_distrito['precio']
    
    # Dividir datos en entrenamiento y prueba
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=27)
    
    # Crear y entrenar el modelo
    modelo = LinearRegression()
    modelo.fit(X_train, Y_train)

    # Predecir valores
    Y_pred = modelo.predict(X_test)
    
    # Graficar los puntos de entrenamiento
    plt.scatter(X_train, Y_train, color=colores_punto[i], label=f'Datos de Entrenamiento: {distrito}', alpha=0.5)

    # Graficar los puntos de prueba
    plt.scatter(X_test, Y_test, color=colores_punto[i], label=f'Datos de Prueba: {distrito}', alpha=0.5)
    
    # Graficar la línea de regresión
    plt.plot(X_train, modelo.predict(X_train), color=colores_linea[i])
    
    # Obtener los coeficientes w y b
    w = modelo.coef_[0]
    b = modelo.intercept_
    
    # Almacenar los coeficientes en el diccionario
    coeficientes_distritos[distrito] = {'w': w, 'b': b}

# Configurar el gráfico
plt.xlabel('Área (m2)')
plt.ylabel('Precio (S/.)')
plt.title('Regresión Lineal por Distrito')
plt.legend()
plt.show()

# Mostrar los coeficientes por distrito
for distrito, coefs in coeficientes_distritos.items():
    print(f"Distrito: {distrito}")
    print(f"  Coeficiente (w): {coefs['w']}")
    print(f"  Intercepto (b): {coefs['b']}")
    print()
```

### Crítica
En la actualidad, el avance de la Inteligencia Artificial (IA) ha permitido el desarrollo de herramientas innovadoras para enfrentar los problemas ambientales que afectan a Perú, un país vulnerable a los desastres naturales. Los proyectos de prevención son fundamentales, ya que apoyan a las autoridades en la toma de medidas preventivas, evitando daños mayores y preparando a la población. Un ejemplo de esto es el proyecto del Instituto Geofísico del Perú, que utiliza algoritmos de aprendizaje automático para predecir con mayor precisión fenómenos como El Niño Global y El Niño Costero. Por otro lado, se están desarrollando proyectos enfocados en la respuesta ante desastres, los cuales buscan mejorar la capacidad de actuar rápidamente una vez que han ocurrido los eventos. Asimismo, se están desarrollando modelos de IA para la previsión meteorológica, que permitirán identificar patrones climáticos con mayor antelación y precisión, mejorando tanto la prevención como la respuesta.

Estos proyectos ofrecen grandes beneficios si se implementan de manera adecuada, como una mejor predicción y respuesta rápida ante desastres naturales, lo que permitiría salvar vidas y minimizar el daño ambiental.
