6. Agrega una nueva columna 'precio_por_area' que calcule el precio por metro cuadrado (precio / area). (1 pts)

```python
# tu código aquí
df_area = pd.DataFrame(data)

df_area['area'] = df_area['area'].str.replace(' m2','').astype(float)

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)
    
df_area['precio_soles'] = df_area['precio'].apply(convertirSoles)
df_area['precio_por_area'] = df_area['precio_soles'] / df_area['area']
df_area['precio_por_area'] = df_area['precio_por_area'].round(2)

df_area
```

7. Agrega una columna `moneda` donde se indique la moneda en la que se encuentra el precio. (0.5pts)

```python
df_moneda = pd.DataFrame(data)

def determinarMoneda(precio):
    if '$' in precio:
        return 'Dólares'
    elif 'S/' in precio:
        return 'Soles'
    else:
        return 'N/A'
    
df_moneda['moneda'] = df_moneda['precio'].apply(determinarMoneda)
df_moneda
```

8. Agrupa el dataset por `distrito` y calcula el precio promedio por distrito. (1 pts)

```python
# tu código aquí
df_distrito = pd.DataFrame(data)

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)


df_distrito['precio'] = df_distrito['precio'].apply(convertirSoles)

precio_promedio_por_distrito = df_distrito.groupby('distrito')['precio'].mean().reset_index()
precio_promedio_por_distrito.rename(columns={'precio': 'precio_promedio'}, inplace=True)
precio_promedio_por_distrito['precio_promedio'] = precio_promedio_por_distrito['precio_promedio'].round(2)


precio_promedio_por_distrito
```

9. ¿Cuál es precio promedio de los departamentos, por distrito, de tipo `flat` y que tienen `gimnasio` como `areas_comunes`? La tabla de salida debe estar ordenada alfabéticamente por el nombre del distrito. (1 pts)

```python
df_depas = pd.DataFrame(data)

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)

df_depas['precio'] = df_depas['precio'].apply(convertirSoles)

df_depas['tiene_gimnasio'] = df_depas['areas_comunes'].apply(lambda x: 'Gimnasio' in x)
df_filtrado_depa = df_depas[(df_depas['tipo'] == 'flat') & (df_depas['tiene_gimnasio'])]
precio_distrito = df_filtrado_depa.groupby(['distrito', 'tipo', 'tiene_gimnasio'])['precio'].mean().reset_index()
precio_distrito.rename(columns={'precio': 'precio_promedio'}, inplace=True)
precio_distrito['precio_promedio'] = precio_distrito['precio_promedio'].round(2)
precio_distrito = precio_distrito.sort_values(by='distrito')

precio_distrito
```

10. A partir del dataset y las columnas, utilizar un modelo de Regresión Linear de ScikitLearn para encontrar el modelo que pueda explicar la correlación entre `precio` y `area`. (4 pts)

Considerando la siguiente ecuación
$Y = X*w + b$

$Y$: `valor del precio`
$X$: `tamaño del departamento en m2` 

$w$ y $b$: coeficientes 


```python
#tu código aquí
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df_depas = pd.DataFrame(data)

#Tratado de datos

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)

df_depas['precio'] = df_depas['precio'].apply(convertirSoles).astype(int)

df_depas['area'] = df_depas['area'].str.replace(' m2','',case=False).astype(float)

simpledf = df_depas[['precio', 'area']]

#modelo de regresion
X = simpledf['area']
Y = simpledf['precio']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.05,random_state=2)

X_train_reshaped = X_train.values.reshape(-1,1)
Y_train_reshaped = Y_train.values.reshape(-1,1)

X_test_reshaped = X_test.values.reshape(-1,1)

model = LinearRegression()
model.fit(X_train_reshaped, Y_train_reshaped)

Y_pred = model.predict(X_test_reshaped)

plt.scatter(X_train, Y_train, color='green')
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_train, model.predict(X_train_reshaped), color= 'yellow')
plt.xlabel('area por m2')
plt.ylabel('precio en S/.')
plt.show()
```

```python
#imprime tus parámetros w y b aquí
# Obtener el coeficiente (w)
w = model.coef_[0]

# Obtener la intersección (b)
b = model.intercept_

print(f'w: {w}')
print(f'b: {b}')
```
