6. Agrega una nueva columna 'precio_por_area' que calcule el precio por metro cuadrado (precio / area). (1 pts)

```python
df_area = pd.DataFrame(data)

df_area['area'] = df_area['area'].str.replace(' m2','').astype(float)

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)
    
df_area['precio_soles'] = df_area['precio'].apply(convertirSoles)
df_area['precio_por_area'] = df_area['precio_soles'] / df_area['area']
df_area['precio_por_area'] = df_area['precio_por_area'].round(2)

df_area
```

10. A partir del dataset y las columnas, utilizar un modelo de Regresión Linear de ScikitLearn para encontrar el modelo que pueda explicar la correlación entre `precio` y `area`. (4 pts)

Considerando la siguiente ecuación
$Y = X*w + b$

$Y$: `valor del precio`
$X$: `tamaño del departamento en m2` 

$w$ y $b$: coeficientes 


```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df_depas = pd.DataFrame(data)

#Tratado de datos

tasa_cambio = 3.7

def convertirSoles(precio):
    if '$' in precio:
        return float(precio.replace('$ ', '').replace(',', '')) * tasa_cambio
    elif 'S/' in precio:
        return float(precio.replace('S/ ', '').replace(',', ''))
    else:
        return float(precio)

df_depas['precio'] = df_depas['precio'].apply(convertirSoles).astype(int)

df_depas['area'] = df_depas['area'].str.replace(' m2','',case=False).astype(float)

simpledf = df_depas[['precio', 'area']]

#modelo de regresion
X = simpledf['area']
Y = simpledf['precio']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.05,random_state=27)

X_train_reshaped = X_train.values.reshape(-1,1)
Y_train_reshaped = Y_train.values.reshape(-1,1)

X_test_reshaped = X_test.values.reshape(-1,1)

model = LinearRegression()
model.fit(X_train_reshaped, Y_train_reshaped)

Y_pred = model.predict(X_test_reshaped)

plt.scatter(X_train, Y_train, color='green')
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_train, model.predict(X_train_reshaped), color= 'yellow')
plt.xlabel('area por m2')
plt.ylabel('precio en S/.')
plt.show()
```

```python
#Imprime tus parámetros w y b aquí

# Obtener el coeficiente (w)
w = model.coef_[0]

# Obtener la intersección (b)
b = model.intercept_

print(f"Coeficiente (w): {w}")
print(f"Intercepto (b): {b}")
```

11.  ¿Cómo varía los coeficientes `w` y `b` dependiendo del distrito? Seleccione 5 casos y analice `w` y `b`. (2 pts)

```python
# Seleccionar 5 distritos
distritos_seleccionados = ['La Victoria', 'Santiago De Surco', 'Miraflores', 'Lince', 'San Miguel']

# Crear un diccionario para almacenar los coeficientes de cada distrito
coeficientes_distritos = {}

# Iterar sobre los distritos seleccionados y entrenar un modelo de regresión para cada uno
for distrito in distritos_seleccionados:
    # Filtrar los datos para el distrito actual
    df_distrito = df_depas[df_depas['distrito'] == distrito]
    
    # Definimos variables
    X = df_distrito['area'].values.reshape(-1, 1)  # Usar datos filtrados
    Y = df_distrito['precio'].values
    
    # Dividir datos en entrenamiento y prueba
    X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.05,random_state=27)
    
    # Crear y entrenar el modelo
    modelo = LinearRegression()
    modelo.fit(X_train, Y_train)
    predicciones = modelo.predict(X_test)
    
    plt.scatter(X_train, Y_train, color='green', label='Entrenamiento')
    plt.scatter(X_test, Y_test, color='red', label='Prueba')
    plt.plot(X_train, modelo.predict(X_train), color='yellow', label='Línea de regresión')
    plt.xlabel('Área (m2)')
    plt.ylabel('Precio (S/.)')
    plt.title(f'Regresión Lineal para {distrito}')
    plt.legend()
    plt.show()

    # Obtener los coeficientes w y b
    w = modelo.coef_[0]
    b = modelo.intercept_
    
    # Almacenar los coeficientes en el diccionario
    coeficientes_distritos[distrito] = {'w': w, 'b': b}

# Mostrar los coeficientes por distrito
for distrito, coefs in coeficientes_distritos.items():
    print(f"Distrito: {distrito}")
    print(f"  Coeficiente (w): {coefs['w']}")
    print(f"  Intercepto (b): {coefs['b']}")
    print()
```
